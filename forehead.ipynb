```python
import pandas as pd
import numpy as np

import torch

from PIL import Image

from glob import glob

import os

from tqdm import tqdm
```

```python
label_frames = pd.read_csv("dataset/label_frame.csv")
label_frames.head()
```

```python
label_array = label_frames.to_numpy()
```

```python
def get_data(data, keys: list, values: list):
    assert len(keys) == len(values), "len(keys) and len(values) must be the same."

    result = []

    for k, v in zip(keys, values):
        for d in data:
            file, forehead, length = d
            if forehead == -1: continue
            elif k == "forehead" and forehead == v:
                result.append([file, forehead, length])
            elif k == "length" and length == v:
                result.append([file, forehead, length])

    return result
```

```python
label_frames = get_data(label_array, keys=["forehead", "forehead", "forehead"], values=['0', '1', '2'])
```

```python
for file, forehead, length in tqdm(label_frames):
    img = Image.open(f"dataset/male/{file}")
    img.save(f"dataset/forehead/{forehead}/{file}")
```
